```python
from sklearn.preprocessing import MinMaxScaler
norm_cars = MinMaxScaler().fit_transform(cars_lr3)
```

```python
norm_cars = pd.DataFrame(norm_cars, columns=cars_lr3.columns)
```

```python
norm_cars
```

```python
# divid to train and test
from sklearn.model_selection import train_test_split        ### normal // correlation -- thinking
```

```python
train, test = train_test_split(norm_cars, test_size=0.3)
```

```python
len(test)
```

```python
train_x = train.drop('price', axis=1)
train_y = train[['price']]

test_x = test.drop('price', axis=1)
test_y = test[['price']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_x, train_y)  #train
```

```python
pred_y = lr.predict(test_x)
```

```python
pred_y
```

```python
print('MSE: ', mean_squared_error(test_y, pred_y))
```
