```python
df2 = df[~df['InvoiceNo'].str.contains('C')]
```

```python
df2.shape
```

```python
df3 = df2[df2['Country'] == 'France']
```

```python
df3.head()
```

```python
df3.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
```

```python
df4 = df3.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
def encoding(x):
    if x >= 1:
        return 1
    else:
        return 0
```

```python
basket_set = df4.applymap(encoding)
```

```python
basket_set.drop(columns=['POSTAGE'], inplace=True)
```

```python
basket_set.head()
```

```python
freq_item = apriori(basket_set, min_support = 0.07, use_colnames=True)
```

```python
df_test = basket_set[basket_set["ALARM CLOCK BAKELIKE GREEN"] + basket_set["ALARM CLOCK BAKELIKE PINK"] == 2]
len(df_test) / len(basket_set)
```

```python
df_test2 = basket_set[basket_set["ALARM CLOCK BAKELIKE GREEN"] + basket_set["ALARM CLOCK BAKELIKE PINK"] == 1]
len(df_test2) / len(df_test)
```
